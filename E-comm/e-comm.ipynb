###  Обработка явных и неявных дубликатов

Сначала посмотрим, есть ли явные повторы.

```python
data.duplicated().sum()
```

Явных повторов не обнаружено. Посмотрим, какие могут быть невяные дубликаты. Например, может задвоится выгрузка данных на разные даты. Проверим дубликаты по ключевым параметрам: клиент, заказ, товар.

```python
duplicateRows = data.duplicated(subset=['customer_id', 'product', 'order_id']).sum()
print(f'Обнаружено {duplicateRows} дубликатов.')
```

Удалим их.

```python
# удаление
data = data.drop_duplicates(subset=['customer_id', 'product', 'order_id'])
```

Еще может быть под одним заказом несколько клиентов. Проверим, есть ли такие случаи.

```python
data.groupby('order_id')['customer_id'].nunique().value_counts()
```

Есть 26 заказов, которые присвоили двум клиентам и 2 одинаковых номера заказа, который присвоили трем клиентам и даже есть заказ, который у четырех клиентов. Такого быть не должно, т.к. в датасете должно быть один клиент - один заказ. Посмотрим на них.

```python
# таблица с заказами и количеством привязанных id клиентов
repeated_orders = data.groupby('order_id')['customer_id'].nunique().sort_values(ascending = False).head(29)
repeated_orders = repeated_orders.reset_index()

# список повторяющихся заказов
repeated_orders_list = repeated_orders['order_id'].to_list()

# таблица несколько клиентов на один заказ
data.query('order_id in @repeated_orders_list').sort_values(by='order_id', ascending = False).head(15)
```

Видим, что повторяется не только номер заказа, но и наименование товара, различаются даты и клиент. У нас таблица должна быть построена по принципу один заказ - один клиент. Повторяющиеся заказы у разных клиентов является ошибкой. Удалим их.

```python
# удаление неявных дубликатов
data = data.query('order_id not in @repeated_orders_list')
```

## Сегментация клиентов

Для сегментации клиентов используем rfm-анализ, чтобы разделить клиентов на разные сегменты.  

RFM означает новизну, частоту, денежную стоимость.   

Недавность: как давно клиент совершил с нами транзакцию.  
Частота: как часто клиент заказывает/покупает какой-либо продукт у нас.  
Денежный: сколько клиент тратит на покупку продуктов у нас.

### Вычисление параметра R recency — давность (как давно ваши клиенты что-то покупали)

Рассчитаем давность для клиентов, совершивших покупку в компании. Так как датасет не самый свежий, вместо текущей даты будем использовать максимальную дату в датасете.

```python
# вычислим последнюю дату покупки каждого клиента
recency = data.groupby('customer_id', as_index=False)['date'].max()
recency.columns = ['customer_id', 'last_date']

# последняя дата в датасете
recent_date = data['date'].max() + timedelta(days=1)

# вычисление кол-ва дней с последней покупки клиента до последней даты датасета
recency['recency'] = recency['last_date'].apply(lambda x: (recent_date - x).days)

recency.head()
```

Показатель `recency` расчитан. Он показывает сколько прошло дней с момента последней покупки клиента до последней даты в датасете.

### Вычисление параметра F - frequency — частота (как часто клиенты покупают)

Рассчитываем частоту частых транзакций клиента при заказе/покупке какого-либо продукта у компании.

```python
# группировка по id заказчика, подсчет количества сделанных заказов
frequency = data.groupby('customer_id', as_index=False)['order_id'].nunique()
frequency.columns = ['customer_id', 'frequency']
frequency.head()
```

Показатель `frequency` показывает сколько покупок делал клиент за рассматриваемый период.

### Вычисление параметра M - monetary — деньги (общая сумма покупок)

Рассчитаем денежную стоимость расходов клиентов на покупку продуктов у компании. Этот параметр был рассчитан в исследовательской анализе.

```python
# выделим необходимые столбцы
monetary = customer_revenue.copy()
monetary.columns = ['customer_id', 'monetary']
monetary.head()
```

Показатель `monetary` показывает выручку клиента за все время.

### Объединение всех трех столбцов в один набор данных

Объединим все столбцы фрейма данных в один объект, используя функцию слияния для отображения давности, частоты, денежного значения.

```python
# объединение recency и frequency
rf = recency.merge(frequency, on='customer_id')
# объединение с monetary
rfm = rf.merge(monetary, on='customer_id').drop(columns='last_date')
rfm.head()
```

Посмотрим на характеристики полученных столбцов.

### Оценки клиентов на основе их давности, частоты и денежной оценки

Теперь нам нужно выставить оценки. Мы будем разбивать на диапазон от 1 до 3.  Присвоим следующие оценки  

Для параметра `recency`:  
- 3 - недавние клиенты  
- 2 - спящие  
- 1 - давние  

Для параметра `frequency`:
- 3 - частые
- 2 - редкие
- 1 - разовые

Для параметра `monetary`:  
- 3 - высокий чек
- 2 - средний чек
- 1 - низкий чек

Для разбивки будем использовать равные промежутки, т.е. 33.3, 66.7 процентили.

```python
# создание функции для сегментации
def r_score(value):
    if value <= np.percentile(rfm['recency'], 33.3):
        return 3
    elif value <= np.percentile(rfm['recency'], 66.7):
        return 2
    else:
        return 1

def fm_score(value, column):
    if value <= np.percentile (rfm[column], 33.3):
        return 1
    elif value <= np.percentile (rfm[column], 66.7):
        return 2
    else:
        return 3
```

```python
# создание столбцов с оценками
rfm['r_score'] = rfm['recency'].apply(lambda x: r_score(x))
rfm['f_score'] = rfm['frequency'].apply(lambda x: fm_score(x, 'frequency'))
rfm['m_score'] = rfm['monetary'].apply(lambda x: fm_score(x, 'monetary'))
rfm
```

Оценки по трем параметрам выставлены. Следущим шагом объединим оценки и присвоим им категории.

### Оценка клиента на основе оценки RFM

Групп по клиентам получилось много и они по численности небольшие. Для удобства разработки предложений для каждой группы, укрупним группы.

Разделим всех клиентов на следующие группы:

**Лояльные клиенты** ( 333, x23 ): покупатели с высокой и средней частотой активности с высоким чеком, независимо от их недавности. Им стоит предложить программы лояльности или специальные акции, чтобы поддерживать их интерес.

**Перспективные клиенты** (32х, 22х): покупатели с высокой и средней частотой активности с любым чеком. Их необходимо привлечь так, чтобы они превратились в лояльных.  

**Клиенты, требущие внимания** (31х, 13х, 23х): покупатели, которые покупали давно, но часто и по какой-то причине перестали покупать, а также клиенты, которые покупали недавно, но редко или всего один раз, необходимо предложить похожий товар, чтоб он заинтересовался и вернулся.  

**Спящие**: все остальные клиенты, ,последняя покупка была давно и мало заказов, могут быть потеряны.

```python
def assign_segment(row): 
    rfm = row['rfm_group']
    if (rfm[1] == '2' and rfm[2] == '3') or (rfm == '333'):
        return 'Лояльные' 
    elif (rfm[0] == '3' and rfm[1] == '2') or  (rfm[0] == '2' and rfm[1] == '2'):
        return 'Перспективные' 
    elif (rfm[0] == '3' and rfm[1] == '1') or  (rfm[0] == '1' and rfm[1] == '3') or rfm[0] == '2' and rfm[1] == '3':
        return 'Требуют внимания' 
    else:
        return 'Спящие'
    
rfm['segment'] = rfm.apply(assign_segment, axis=1) 

rfm
```

Все клиенты распределены по сегментам. Посмотрим, какое количество клиентов в каждом сегменте.

Для более легкого просмотра категорий, выделим и оставим первые два слова в названии товара.

## Проверка гипотезы

Вернемся к сегментации покупателей. Необходимо проверить, действительно ли группы отличаются друг от друга. Для сравнительной характеристики будем использовать выручку с покупателя, средний чек клиента и давность заказа.

### Гипотеза о выручка с покупателя

Сформируем гипотезы:  
H0 - Нет различий между средними выручками покупателей  
H1 - Есть различия между средними выручками покупателей  

Для того, чтобы определить метод проверки гипотез, определим, нормально ли распределeны значения по выручке с помощью критерия Шапиро-Уилка.

```python
alpha_norm = 0.05  # критический уровень статистической значимости

results = st.shapiro(customer_revenue['revenue'])
p_value = results[1]  # второе значение в массиве результатов (с индексом 1) - p-value

print('p-значение: ', p_value)

if p_value < alpha_norm:
    print('Отвергаем нулевую гипотезу: распределение не нормально')
else:
    print('Не получилось отвергнуть нулевую гипотезу, всё нормально') 
```

Данные по выручке - количественные, распределены не нормально, группы покупателей не зависят друг от друга по этому параметру, сравнивать будем по две группы. Для проверки гипотез подходит критерий Манна-Уитни.

Подготовим данные. Выделим в отдельную таблицу заказчик - сегмент, к которому он относится и выручка с каждого заказчика.

```python
tabl_for_hyp = rfm.merge(customer_revenue, on='customer_id')[['customer_id', 'segment', 'revenue']]
tabl_for_hyp
```

Сохраним таблицы для каждого сегмента

```python
loyal = tabl_for_hyp.query('segment == "Лояльные"')[['revenue']]
perspective = tabl_for_hyp.query('segment == "Перспективные"')[['revenue']]
sleepy = tabl_for_hyp.query('segment == "Спящие"')[['revenue']]
attention = tabl_for_hyp.query('segment == "Требуют внимания"')[['revenue']]
```

```python
# функция для проверки гипотез
def stat_hyp_revenue (group_1, group_2, name_group_1, name_group_2, alpha):
    results = st.mannwhitneyu(group_1, group_2)
    alpha = alpha

    print(f'Гипотеза H0: Выручка с клиента группы {name_group_1} = выручке с клиента группы {name_group_2}')
    print(f'Гипотеза H1: Выручка с клиента группы {name_group_1} ≠ выручке с клиента группы {name_group_2}')
    
    print(results.pvalue)

    if results.pvalue < alpha:
        print('Отвергаем нулевую гипотезу') 
    else:
        print('Не получилось отвергнуть нулевую гипотезу')
    print('________________________________')
```

Укажем порог статистической значимости в 5%. Однако, т.к. проверка является множестенной введем поправку Бонферрони.

```python
# вероятность ошибки
alpha_1 = 0.05 # уровень принимаемой статистической значимости
n = 6 # количество тестируемых гипотез
mist = round(1-(1-alpha_1)**n,2)
print('Принятая статистическая значимость:', alpha_1)
print('Вероятность ошибки:', mist)
```

Вероятность ошибки (0.26) больше принятого уровня статистической значимости (0.05). Значит, выбранный уровень значимости необходимо пересчитать. Для этого используем поправку Бонферрони.

```python
# поправка Бонферрони
m = 6 #количество гипотез
alpha = alpha_1 / m
```

```python
# вероятность ошибки
mist_correct = round(1-(1-alpha)**n,2)
print('Принятая статистическая значимость:', alpha)
print('Вероятность ошибки:', mist_correct)
```

Вероятность ошибки стала равна уровню статистистической значимости, а значит, примем ее значение равной 0.008333333333333333.

```python
stat_hyp_revenue (loyal, perspective , 'loyal', 'perspective', alpha)
stat_hyp_revenue (loyal, sleepy , 'loyal ', 'sleepy', alpha)
stat_hyp_revenue (loyal, attention , 'loyal ', 'attention', alpha)

stat_hyp_revenue (perspective, sleepy , 'perspective ', 'sleepy', alpha)
stat_hyp_revenue (perspective, attention , 'perspective ', 'attention', alpha)

stat_hyp_revenue (sleepy, attention , 'sleepy ', 'attention', alpha)
```

В группах `perspective` - `attention` и `sleepy`- `attention` выручки оказались равны. Проверим, возможо, эти группы будут отличаться по другому параметру.
